```python
plt.hist(x , bins=3)
plt.xlabel('TV')
plt.show
```

```python
plt.hist(x['TV'] , bins=10)
plt.xlabel('TV')
plt.title('TV histogram')
plt.show
```

```python
plt.hist(x['Radio'] , bins=10 , color='green')
plt.xlabel('Radio')
plt.title('Radio histogram')
plt.show
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=x[['TV', 'Radio', 'Newspaper', 'Sales']], palette='Set2')
plt.xlabel('Features')
plt.ylabel('Values')
plt.title('Box Plots of TV, Radio, Newspaper, and Sales')
plt.show()
```

```python
plt.figure(figsize=(14, 10))
plt.subplot(2, 2, 1)
plt.hist(x['TV'], bins=20, color='blue', edgecolor='black')
plt.title('TV Distribution')
plt.xlabel('TV')
plt.ylabel('Frequency')
plt.subplot(2, 2, 2)
plt.hist(x['Radio'], bins=20, color='red', edgecolor='black')
plt.title('Radio Distribution')
plt.xlabel('Radio')
plt.ylabel('Frequency')
plt.subplot(2, 2, 3)
plt.hist(x['Newspaper'], bins=20, color='green', edgecolor='black')
plt.title('Newspaper Distribution')
plt.xlabel('Newspaper')
plt.ylabel('Frequency')
plt.subplot(2, 2, 4)
plt.hist(x['Sales'], bins=20, color='purple', edgecolor='black')
plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(x['TV'], bins=20, kde=True, color='blue')
plt.title('Histogram and Density Plot of TV Advertising')
plt.xlabel('TV Advertising Spend')
plt.ylabel('Frequency')
plt.subplot(1, 2, 2)
sns.boxplot(x['TV'], color='lightgreen')
plt.title('Box Plot of TV Advertising Spend')
plt.xlabel('TV Advertising Spend')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(x['Radio'], bins=20, kde=True, color='green')
plt.title('Histogram and Density Plot of Radio Advertising')
plt.xlabel('Radio Advertising Spend')
plt.ylabel('Frequency')
plt.subplot(1, 2, 2)
sns.boxplot(x['Radio'], color='yellow')
plt.title('Box Plot of Radio Advertising Spend')
plt.xlabel('Radio Advertising Spend')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(x['Newspaper'], bins=20, kde=True, color='purple')
plt.title('Histogram and Density Plot of Newspaper Advertising')
plt.xlabel('Newspaper Advertising Spend')
plt.ylabel('Frequency')
plt.subplot(1, 2, 2)
sns.boxplot(x['Newspaper'], color='green')
plt.title('Box Plot of Newspaper Advertising Spend')
plt.xlabel(' Newspaper Advertising Spend')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(x['Sales'], bins=20, kde=True, color='red')
plt.title('Histogram and Density Plot of Sales Advertising')
plt.xlabel('Sales Advertising Spend')
plt.ylabel('Frequency')
plt.subplot(1, 2, 2)
sns.boxplot(x['TV'], color='lightgreen')
plt.title('Box Plot of Sales Advertising Spend')
plt.xlabel('Sales Advertising Spend')
plt.tight_layout()
plt.show()
```
